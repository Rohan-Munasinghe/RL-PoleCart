```python
import os
import gym
import random
import time
!pip install pygame
```

```python
from stable_baselines3 import PPO # one of the algorithms
from stable_baselines3.common.vec_env import DummyVecEnv # wrapper around the environment
from stable_baselines3.common.evaluation import evaluate_policy # allows to get mean reward and std dev over an epoch
```

use OpenAi's CartPole-v1 environment and "make" method of the gym library to instantiate the "env"

```python
environment_name = 'CartPole-v1'
env = gym.make('CartPole-v1',render_mode='human')
states = env.observation_space.shape[0] # read states x,theta, xdot, thetadot
actions = env.action_space.n # read actions - left, right Discrete(2)
```

Silulmate the Environment for randome input

```python
episodes = 10
for episode in range(1, episodes+1): # for 10 episodes
    state = env.reset() # Observations of the environment
    done = False
    score = 0 # each episode starts with 0 reward

    while not done: # for each episode
        env.render() # show graphics
        action = random.choice([0,1]) # seelct left or right push randomely
        result = env.step(action) # apply action
        n_state = result[0]
        reward = result[1]
        done = result[2]
        info = result[3]
        score += reward # accumulate reward
    print('Episode:{}, score:{}'.format(episode,score))
    time.sleep(0)
env.close()
```

Save the model in the training folder inside Saved_models by the name PPO_Model_Cartpole

```python
log_path = os.path.join('Training','Logs') # You have to have the folders 'Training' and inside it 'Logs'
```

```python
env = gym.make(environment_name)
env = DummyVecEnv([lambda: env]) # Encapsulate the environment with DummyVector wrapper
model = PPO('MlpPolicy', env, verbose=1, tensorboard_log=log_path) # Model used for trainng in PPO (Proximal Policy Optimization)
# PPO?? # print hyper-paraeters in the model/Algorithm
```

```python
model.learn(total_timesteps=20000)
```

Save Trained Model

```python
PPO_Path = os.path.join('Training','Saved_models','PPO_Model_Cartpole')
```

```python
model.save(PPO_Path)
```

```python
del model
model = PPO.load(PPO_Path, env=env) # Reload the model from its location to the environment env
```

Evaluate Trained Model

```python
evaluate_policy(model, env, n_eval_episodes=10)
```

Insert Callback for training

```python
from stable_baselines3.common.callbacks import EvalCallback,StopTrainingOnRewardThreshold
```

```python
save_path = os.path.join('Training','Saved Models')
```

Create a Callback for learning

```python
stop_callback = StopTrainingOnRewardThreshold(reward_threshold=500, verbose=1)
eval_callback = EvalCallback(env,
                             callback_on_new_best=stop_callback,
                             eval_freq=10000,
                             best_model_save_path=save_path,
                             verbose=1)
```

Create a new Model (for trining with callback)

```python
env = gym.make(environment_name)
env = DummyVecEnv([lambda: env]) # Encapsulate the environment with DummyVector wrapper
model = PPO('MlpPolicy', env, verbose=1, tensorboard_log=log_path) # Model used for trainng in PPO
```

```python
model.learn(total_timesteps=30000, callback=eval_callback) # learning through 20k action steps with eval_callback
```

Create a new NN model

```python
net_arch = [dict(pi=[128,128,128,128], vf=[128,128,128,128])] # four layers each having 128 nodes
model = PPO('MlpPolicy', env, verbose=1, tensorboard_log=log_path, policy_kwargs={'net_arch':net_arch})
```

```python
model.learn(total_timesteps=30000, callback=eval_callback) # learning through 20k action steps with eval_callback
```
